```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Gas data

```python
gas = pd.read_csv("gas_prices.csv")
```

```python
display(gas)
```

```python
plt.figure(figsize=(8, 5))

plt.title("Gas Prices over Time (in USD)", fontdict={"fontweight": "bold"})

# countries_lst = gas.columns[1:]
# for country in countries_lst:
#     plt.plot(gas.Year, gas[country], label=country, marker=".")

plt.plot(gas.Year, gas.USA, "b.-", label="USA")
plt.plot(gas.Year, gas.Canada, "g.-", label="Canada")
plt.plot(gas.Year, gas["South Korea"], "y.-", label="South Korea")
plt.plot(gas.Year, gas.Australia, "r.-", label="Australia")

plt.xlabel("Year")
plt.ylabel("US Dollars")

plt.xticks(gas.Year[::3].to_list() + [2011])

plt.legend()

plt.savefig("gas_price_fig.png", dpi=300)

plt.show()
```

## Fifa data

```python
fifa = pd.read_csv("fifa_data.csv")
```

```python
fifa.head()
```

### Histograms

```python
bins = [40, 50, 60, 70, 80, 90, 100]

plt.figure(figsize=(8, 5))

plt.title("Distribution of player skills in FIFA game")
plt.hist(fifa.Overall, bins=bins, color="#9da9e0")

plt.xticks(bins)

plt.xlabel("Skill level")
plt.ylabel("N° players")

plt.show()
```

### Pie

```python
left = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0]
right = fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]

plt.title("Preferred Foot")

labels = ["Left", "Right"]
colors = ["#abcdef", "#aabbcc"]
plt.pie([left, right], labels=labels, colors=colors, autopct="%.2f %%")

plt.show()
```

```python
fifa.Weight = [int(x.strip("lbs")) if type(x)== str else x for x in fifa.Weight]
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ["Under 125", "125-150", "150-175", "175-200", "Over 200"]
explode = (.4, .1, 0, 0, .4)

plt.style.use("ggplot")

plt.title("Weight distribution of FIFA players in lbs")

plt.pie(weights, labels=labels, autopct="%.2f %%", pctdistance=0.95, explode=explode)

plt.show()
```

```python
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]["Overall"]
madrid = fifa.loc[fifa.Club == "Real Madrid"]["Overall"]
revs = fifa.loc[fifa.Club == "New England Revolution"]["Overall"]

plt.style.use("default")

plt.figure(figsize=(5, 10))
labels = ["Barcelona", "Real Madrid", "Revs"]

plt.title("Professional Soccer Team Comparison")
                                                                # facecolor          # median propertie
boxes = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True, medianprops={"linewidth": 2})

for i, box in enumerate(boxes["boxes"]):
    box.set(color="#4286f4", linewidth=i + 0.5)
    box.set(facecolor="#e0e0e0")

plt.ylabel("FIFA overall rating")

plt.show()
```
