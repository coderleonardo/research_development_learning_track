```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x, y, label="2x",
                 color="red", linewidth=3, marker=".",
                  markersize=15, markeredgecolor="blue", 
                  linestyle="--")

plt.title("First graph!!!", fontdict={"fontname": "Oswald", "fontsize": 20})
plt.xlabel("X label")
plt.ylabel("Y label")

plt.xticks([1, 2, 3, 4])
plt.yticks([0, 1, 2, 3, 4, 5, 6, 7])

plt.legend()


plt.show()
```

```python
x = [1, 2, 3]
y = [2, 4, 6]

# Resize
plt.figure(figsize=(3, 2), dpi=100)

## Line 1
# shorthand notation
plt.plot(x, y, "b^--", label="2x")

plt.title("First graph!!!", fontdict={"fontname": "Oswald", "fontsize": 20})
plt.xlabel("X label")
plt.ylabel("Y label")

## Line 2
x2 = np.arange(0, 4, 0.5)
plt.plot(x2[:6], x2[:6]**2, "r", label="x^2")
plt.plot(x2[5:], x2[5:]**2, "r--")

plt.xticks([1, 2, 3, 4])
#plt.yticks([0, 1, 2, 3, 4, 5, 6, 7])

plt.legend()

plt.savefig("mygraph.png", dpi=300)


plt.show()
```

### Bar Chart 

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]

plt.figure(figsize=(6, 4))

bars = plt.bar(labels, values)

patterns = ["/", "O", "*"]
for bar in bars:
    bar.set_hatch(patterns.pop(0))
# bars[0].set_hatch("/")
# bars[1].set_hatch("O")
# bars[2].set_hatch("*")


plt.show()
```
