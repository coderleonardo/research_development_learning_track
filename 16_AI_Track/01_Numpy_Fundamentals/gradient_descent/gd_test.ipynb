```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.plot(X, y, "b.")
plt.show()
```

```python
from gradient_descent import LBGD, LSGD
```

```python
lr_bgd = LBGD()
lr_bgd.fit(X, y)

y_pred = lr_bgd.predict(X)

print(lr_bgd.theta)

plt.plot(X, y_pred, "r-")
plt.plot(X, y, "b.")
plt.show()
```

```python
lr_sgd = LSGD()
lr_sgd.fit(X, y)

y_pred = lr_sgd.predict(X)

print(lr_sgd.theta)

plt.plot(X, y_pred, "r-")
plt.plot(X, y, "b.")
plt.show()
```
